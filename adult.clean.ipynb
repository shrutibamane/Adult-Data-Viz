```python
import pandas as pd
import numpy as np
```

```python
adult_data= 'adult.data'
adult_test = 'adult.test'
cols = ['age','workclass','final_weight','education','education-num','marital-status',
        'occupation','relationship','race','sex','capital-gain', 'capital-loss',
        'hours-per-week', 'native-country','compensation']
df = (pd.read_csv(adult_data,
                 names=cols,
                 sep=', ',
                 engine='python')
     .append(pd.read_csv(adult_test,
                         skiprows=1,
                         names=cols,
                         sep=', ',
                         engine='python')))

df.head()
```

```python
df = df.replace({'<=50K.' : '<=50K',
               '>50K.'  : '>50K'})
```

```python
df.loc[df['occupation']=='Armed-Forces','occupation'] = 'Protective-serv'
```

```python
df = (df[(df['workclass']!='?')&
       (df['occupation']!='?')&
       (df['workclass']!='Without-pay')]
     .reset_index(drop=True))
df['idx'] = df.index
```

```python
df.loc[df['workclass'].isin(['State-gov', 'Federal-gov', 'Local-gov']), 
      'employment-type'] = 'Government'
df.loc[df['workclass'].isin(['Self-emp-not-inc', 'Self-emp-inc']),      
      'employment-type'] = 'Self-Employed'
df.loc[df['workclass'].isin(['Private']),                               
      'employment-type'] = 'Privately-Employed'
df['employment-type'].unique()
```

```python
df.loc[df['education-num'] <= 8,          'education'] = 'Less than High School'
df.loc[df['education-num'].isin([ 9,10]), 'education'] = 'High School'
df.loc[df['education-num'].isin([11,12]), 'education'] = 'Associates'
df.loc[df['education-num'].isin([13]),    'education'] = 'Bachelors'
df.loc[df['education-num'].isin([14]),    'education'] = 'Masters'
df.loc[df['education-num'].isin([15,16]), 'education'] = 'PhD/Professional'
df['education'].unique()
```

```python
df = df.rename(columns={'hours-per-week':'weekly-hours'})
df.head()
```

```python
df['age raw'] = df['age']
df['employment-type raw'] = df['employment-type']
df['education raw'] = df['education']
df['occupation raw'] = df['occupation']
df['race raw'] = df['race']
df['sex raw'] = df['sex']
df['weekly-hours raw'] = df['weekly-hours']
df['compensation raw'] = df['compensation']
```

```python
df['employment-type'] = df['employment-type'].astype('category')
df['employment-type'] = df['employment-type'].cat.codes
df[['employment-type','employment-type raw','idx']].groupby(['employment-type',
                                                            'employment-type raw']).count()
```

```python
df['education'] = df['education'].astype('category')
df['education'] = df['education'].cat.codes
df[['education','education raw','idx']].groupby(['education',
                                                'education raw']).count()
```

```python
df['occupation'] = df['occupation'].astype('category')
df['occupation'] = df['occupation'].cat.codes
df[['occupation','occupation raw','idx']].groupby(['occupation',
                                                  'occupation raw']).count()
```

```python
df['race'] = df['race'].astype('category')
df['race'] = df['race'].cat.codes
df[['race','race raw','idx']].groupby(['race',
                                      'race raw']).count()
```

```python
df['sex'] = df['sex'].astype('category')
df['sex'] = df['sex'].cat.codes
df[['sex','sex raw','idx']].groupby(['sex',
                                    'sex raw']).count()
```

```python
df['compensation'] = df['compensation'].astype('category')
df['compensation'] = df['compensation'].cat.codes
df[['compensation','compensation raw','idx']].groupby(['compensation',
                                                      'compensation raw']).count()
```

```python
df = df[['idx',
       'age',
       'employment-type',
       'education',
       'occupation',
       'race',
       'sex',
       'weekly-hours',
       'compensation']].copy()

df.head()
```
